```python
from PIL import Image
import os
import glob
from tqdm import tqdm

path = 'F:/open/train_dataset/' # 원본 이미지 경로
new_path = '448size_train/'  # 저장할 이미지 경로

dataset = {}
# resize 시작 --------------------
for where in os.listdir(path): # 품종폴더 BC/LT
    print(where)
    for each in tqdm(os.listdir(path + where)): # 각 품종 번호
        print(each)
        save_path = str(new_path + where + '/' + each)
        print(save_path)
        if not os.path.exists(save_path):
            os.makedirs(save_path)
        for img_path in glob.glob(path + where + '/'+ each + '/' + '*.png'): # 폴더 내에 있는 모든 png 파일들 경로
            img_path = img_path.replace("\\", "/")
            temp = (new_path + img_path[22:])[-10:]
            print(temp)
            im = Image.open(img_path)
            # 이미지 resize
            im = im.resize((448, 448))

            # 이미지 JPG로 저장
            im = im.convert('RGB')
            im.save(save_path + temp)
```
